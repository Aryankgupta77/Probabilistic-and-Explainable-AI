## Open Data Files in pandas
We might have our data in .csv files and if we want to analyze that data using pandas, the first step will be to read it into a data structure that’s compatible with pandas.

### Pandas Data Structures
There are two types of data structures in pandas:
* Series
* DataFrames

#### Pandas Series
A pandas Series is a one-dimensional data structure (“a one-dimensional array”) that stores values, and for every value, it holds a unique index. 

````python
s_list = [1, 7, 2]

print(s_list)
````

````python
p_series = pd.Series(s_list)

print(p_series)
````

#### Pandas Dataframe
A pandas DataFrame is a two-dimensional data structure – basically a table with rows and columns. The columns have names and the rows have indexes. Compared to a pandas Series (which was one labeled column only), a DataFrame is practically the whole data table. You can think of it as a collection of pandas Series (columns next to each other).

Creating DataFrames right in Python is good to know and quite useful when testing new methods and functions you find in the pandas docs.
<br><br>
There are many ways to create a DataFrame from scratch, but a great option is to use a simple dictionary.

````python
data = {
  "calories": [420, 380, 390, 400, 410, 385, 395],
  "grams": [50, 40, 45, 55, 50, 48, 51]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df) 
````

Each (key, value) item in **data** corresponds to a column in the resulting DataFrame. The Index of this DataFrame was given to us on creation as the numbers 0-6, but we could also create our own when we initialize the DataFrame.
<br><br>
Let's have customer names as our index:

````python
daily_portions = pd.DataFrame(data, index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

print(daily_portions)
````

<br><br>

* object - a collection of name/value pairs
* array - an ordered list of values
<br><br>
Object declaration (example):
```python
{ 
        "id": 4911, 
        "firstName":"John",  
        "lastName":"Walter", 
        "isStudent": true

```
<br><br>
Array declaration (example):
```python
[
     "Ireland", "Bulgaria", "UK", "Canada", "India", "China"
```

### Handle Duplicates

It is important to verify that we does not aggregate duplicate rows.

````python
has_duplicates = daily_portions.duplicated().any() 

print(has_duplicates)
````

This dataset does not have any duplicates. To demonstrate duplicates, we will double the records in our DataFrame by appending it to itself.

````python
daily_portions = pd.concat([daily_portions, daily_portions], ignore_index=True)

has_duplicates = daily_portions.duplicated().any() 

print(has_duplicates)
````

````python
daily_portions.shape
````

Drop duplicates.

````python
daily_portions = daily_portions.drop_duplicates()

has_duplicates = daily_portions.duplicated().any() 

print(has_duplicates)

daily_portions.shape
````

Note: The **drop_duplicates()** method returns a copy of our DataFrame. By using the parameter **inplace=True** we will force the dataframe to modify its own object.

````python
daily_portions.drop_duplicates(inplace=True)
````

### Deal with Missing Values

When loading a real-life datasets, most likely there will be missing (or null) values, which are essentially placeholders for non-existent values. In this case, we see **None** or **NaN** values.

There are two options in dealing with nulls:

* Get rid of rows or columns with nulls.
* Replace nulls with non-null values.

````python
url_csv = "https://raw.githubusercontent.com/evassev/CS6501_SEM2_2022/main/Etivities/data/trade_combined.csv"

df = pd.read_csv(url_csv)

df.head(5)
````

````python
df.shape
````

The method **isnull()** returns a DataFrame where each cell is either True or False depending on that cell's null status.

````python
df_null_check = df.isnull()

df_null_check.head(5)
````

To count the number of nulls in each column, we can use an *aggregate function for summing* **sum()**.

````python
df.isnull().sum()
````

#### Removing Null Values
The method **df.dropna()** will delete any row with at least a single null value, but it will return a new DataFrame without altering the original one. To alter the current dataframe, we could specify **inplace=True** in this method.

````python
df.dropna(inplace=True)
````

````python
df.isnull().sum()
````

````python
df.shape
````

We can also drop columns with null values by setting **axis=1**.

````python
df.dropna(axis=1) #drop all the columns with null values
````

### Conditional Selection
We can filter a DataFrame to show only records that comply with certain conditions.

````python
conditional_df = df[df['SecurityType'] == 'PROV']

conditional_df.head(5)
````

````python
conditional_df = df[df['SecurityType'] == 'PROV']

conditional_df = conditional_df[conditional_df['OrderQty'] > 1000000]

conditional_df.head(5)
````

We can make some more complex conditionals by using logical operators **|** for "or" and **&** for "and".

````python
conditional_df = df[(df['SecurityType'] == 'PROV') & (df['OrderQty'] > 1000000)]

conditional_df.head(5)
````

### Apply Functions on Columns 
To iterate over a DataFramea' column, we may use the **apply()** method. For example, we could use a function that checks the order quantity and rank this as "low", "normal" or "high" and use this transformed values to create a new column.

````python
def estimate_qty(x):
    if x <= 1000000:
        return "low quantity"
    elif x <= 1500000:
        return "normal quantity"
    else:
        return "high quantity"
````

Now we send the entire **OrderQty** column through this function by using the **apply()** method.

````python
df['QtyEstimation'] = df['OrderQty'].apply(estimate_qty)

df.head(5)
````

### Convert a Time-Series Data
We convert a time-series data to a series of numbers.

````python
url_csv = 'https://raw.githubusercontent.com/evassev/CS6501_SEM2_2022/main/Lessons/data/relative_price_change_CTtoBS_table.csv'

#df = pd.read_csv('data\\relative_price_change_CTtoBS_table.csv')
df = pd.read_csv(url_csv)

df.head(5)
````

````python
date_unique = df['date'].unique()
class_mapper = {}
num = 0
for element in date_unique:
    num += 1
    class_mapper[element] = num
    
df['date'] = df['date'].replace(class_mapper) 
````
